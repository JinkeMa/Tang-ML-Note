```python
# 复杂步骤构建预测模型
import numpy as np
import torch 
import pandas as pd
import seaborn as sb
import sklearn.preprocessing as preprocessing
import warnings
warnings.filterwarnings('ignore')

features = pd.read_csv('/GitRepos/data/temps.csv')
# onehot code for non-number element
features = pd.get_dummies(features)
features.head()
```

```python
# get labels y
labels = np.array(features['actual'])

# drop colum y
features = features.drop(['year','actual'],axis=1)

feature_list = list(features.columns)
features = np.array(features)

# 特征之间尺度不一致 要进行数据标准化
features = preprocessing.StandardScaler().fit_transform(features)
features.shape
# 构建模型
x = torch.tensor(features,dtype=float)
y = torch.tensor(labels,dtype=float)
```

```python
# device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
weights = torch.randn((13,128), dtype = float, requires_grad = True)
bias = torch.randn(128,dtype = float,requires_grad=True)
weights2 = torch.randn((128,1),dtype=float,requires_grad=True)
bias2 = torch.randn(1,dtype=float,requires_grad=True)
```

```python
lr = 0.01
losses = []

for epoch in range(1000):
    epoch += 1
    # foreward
    hidden = x.mm(weights) + bias
    # activation
    hidden = torch.relu(hidden)
    # prediction
    pred = hidden.mm(weights2) + bias2
    # loss
    loss = torch.mean((pred - y) ** 2)
    # backward
    loss.backward()
    # print
    losses.append(loss)
    if epoch % 50 == 0:
        print('loss {}',loss)
    # update
    weights.data.add_( -1*weights.grad.data*lr)
    bias.data.add_(-1*bias.grad.data)
    weights2.data.add_(-1*weights2.grad.data*lr)
    bias2.data.add_(-1*bias2.grad.data*lr)
    # zero grad
    weights.grad.data.zero_()
    bias.grad.data.zero_()
    weights2.grad.data.zero_()
    bias2.grad.data.zero_()
```
